1. Find the top 1000 restaurants that has rating above 3.5 in a descending order.

```python
%%sql
CREATE TABLE Top_1000 AS
select *
from business
where stars>= 3.5 and categories like "%restaurant%"
order by stars desc
limit 1000;
```

2. We aim to find out which city own the most high rating restaurants.

```python
%%sql
SELECT COUNT(city) AS city_count, city, postal_code
FROM Top_1000
GROUP BY city, postal_code
order by city_count desc;
```

4. Our team have deicided to pick the cuscine type among Korean, Chinese and Japanese. We first use the query to see how many existing Korean restaurants in Santa Barbara.

```python
%%sql
SELECT COUNT(*) AS row_count, name, postal_code, categories
FROM (
  SELECT *
  FROM business
  WHERE city = 'Santa Barbara'
) AS subquery
WHERE categories LIKE '%Korean%'
group by name, postal_code, categories;
```

5. We then write a query to see how existing Chinese restaurants in the market.

```python
%%sql
SELECT COUNT(*) AS row_count, name, postal_code, categories
FROM (
  SELECT *
  FROM business
  WHERE city = 'Santa Barbara'
) AS subquery
WHERE categories LIKE '%Chinese%'
group by name, postal_code, categories;
```

6. At last, we want to check how many exisitng Japanese restaurants in the market.

```python
%%sql
SELECT COUNT(*) AS row_count, name, postal_code, categories
FROM (
  SELECT *
  FROM business
  WHERE city = 'Santa Barbara'
) AS subquery
WHERE categories LIKE '%Japanese%'
group by name, postal_code, categories;
```

7. We have decided to open a Korean restaurant since there are less competitors in the market. As a result, we want to discover the attributes that our current competitors have.

```python
%%sql
SELECT COUNT(*) AS row_count, name, postal_code, categories, attributes, MAX(LENGTH(attributes)) AS max_length
FROM (
  SELECT *
  FROM business
  WHERE city = 'Santa Barbara'
  AND categories LIKE '%Korean%'
) AS subquery
GROUP BY name, postal_code, categories, attributes
ORDER BY max_length DESC;
```

8. We want to decide our opening hours by referencing our competitor's.

```python
%%sql
SELECT COUNT(*) AS row_count, name, postal_code, categories, attributes,hours
FROM (
  SELECT *
  FROM business
  WHERE city = 'Santa Barbara'
  AND categories LIKE '%Korean%'
) AS subquery
GROUP BY name, postal_code, categories, attributes,hours
```

9. In order to learn from our competitors, we would like to check the reviews of the seven restaurants.

```python
%%sql
CREATE TABLE korean AS
SELECT COUNT(*) AS row_count, name, postal_code, categories
FROM (
  SELECT *
  FROM business
  WHERE city = 'Santa Barbara'
) AS subquery
WHERE categories LIKE '%Korean%'
GROUP BY name, postal_code, categories;
```

```python
%%sql
SELECT business_id, text
FROM review
WHERE business_id IN (SELECT business_id FROM korean)
limit 1000;
```

10. By picking the keywords from the reviews we selected from the previous query, we can know more about the customer's appetite.

```python
%%sql
create table koreanreview as
SELECT business_id, text
FROM review
WHERE business_id IN (SELECT business_id FROM korean)
limit 1000;
```

```python
%%sql
select text, business_id
from koreanreview
where text like "%excellent%" or "%good%" or "%fantastic%" or "%great%" or "%amazing%" or "%delicious%" or "%yumy%" or "%fabulous%" or "%love%" or "%like%" or "%awesome%" or "%comfortable%" or "%elegant%" or "%satisfied%" or "%thank%"
```
